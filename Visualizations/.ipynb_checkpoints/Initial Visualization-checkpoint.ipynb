```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso, LassoCV, LinearRegression
from sklearn.model_selection import cross_val_score, train_test_split, KFold
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
from sklearn.preprocessing import LabelEncoder, StandardScaler
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(2)
```

```python
df = pd.read_csv('../Data/kc_final.csv', index_col = 0)
```

# Initial Visualization

```python
df.describe()
```

```python
for var in df:
    if var=="id" or var=="date": 
        continue
    maxVar = max(df[var])
    minVar = min(df[var])
    nBins = int(maxVar-minVar) + 1
    if var in list(["price", "sqft_living", "sqft_lot", "sqft_above", "sqft_basement", "lat", "long", "sqft_living15", "sqft_lot15"]):
        nBins = 100
    if var=="floors" or var=="yr_renovated":
        nBins = df[var].nunique()
    values, bins, bars = plt.hist(df[var], range=[minVar,maxVar], bins=nBins)
    plt.title(var)
    plt.xlabel(var)
    plt.ylabel("Frequency")
    plt.show()
```

```python
for var in df:
    if var=="id" or var=="date" or var=="price": 
        continue
    df.plot(x=var, y="price", style=".")
    plt.title("Correlation between " + var + " and price")
    plt.xlabel(var)
    plt.ylabel("price")
    plt.show()
    pearson = np.corrcoef(df[var], df["price"])
    print(pearson)
```
